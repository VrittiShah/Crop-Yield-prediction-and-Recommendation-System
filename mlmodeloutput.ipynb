```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("finalised_dataset.csv")
```

```python
crop_distribution = df['crop_names'].value_counts(normalize=True) * 100
rare_crop_types = crop_distribution[crop_distribution < 1]
crop_for_removal=crop_distribution[crop_distribution < 0.5]
mask = df['crop_names'].isin(crop_for_removal.index)
df1 = df[~mask]
```

```python
df1 = df1.drop(df1[df1['production'].isnull()].index)
df1 = df1.drop(df1[df1['Yield'].isnull()].index)
```

```python
df=df.drop('state_names',axis=1)
```

```python
df = df1[df1['season_names'] != 'Autumn']
```

```python
df = df[~df['district_names'].isin(['PALGHAR', 'MUMBAI'])]
```

```python
df2 = df.loc[df['crop_year'] != 1997]
```

```python
df=df2
```

```python
df=df.drop(['state_names'],axis=1)
df
```

```python
df = df.drop('production', axis=1)
df=df.drop('area',axis=1)
df
```

```python
numeric_cols = ['crop_year', 'temperature', 'wind_speed', 'pressure', 'humidity', 'N', 'P', 'K', 'Yield']
categorical_cols = ['district_names', 'season_names', 'crop_names', 'soil_type']
ordered_cols = numeric_cols + categorical_cols
df = df[ordered_cols]
df
```

```python
dfx=df
dfx=df.drop('district_names',axis=1)
dfx=dfx.drop('season_names',axis=1)
dfx=dfx.drop('soil_type',axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

# Encode the labels in the specified column
dfx['crop_names'] = encoder.fit_transform(dfx['crop_names'])
```

```python
dfx.head()
```

```python
from sklearn.model_selection import train_test_split
y = dfx['Yield']
X =dfx.drop('Yield', axis = 1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
'''
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
'''
print(X_train)
print(X_test)

from sklearn.ensemble import RandomForestRegressor
regr = RandomForestRegressor(random_state=42, n_estimators=100)
regr.fit(X_train, y_train)
y_pred = regr.predict(X_test)

from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_absolute_error as mae
from sklearn.metrics import r2_score

print('MSE =', mse(y_test, y_pred))
print('MAE =', mae(y_test, y_pred))
print('R2 Score =', r2_score(y_test, y_pred))
```

```python
def prepare_user_input(user_data):

  columns = ['crop_year','temperature','wind_speed','pressure','humidity','N','P','K','crop_names']
  user_df = pd.DataFrame(user_data, columns=columns[:-1], index=[0])
  # user_df = pd.DataFrame(user_data, columns=columns[:-1])  # Exclude crop_names

  le = LabelEncoder()
  user_df['crop_names'] = le.fit_transform([user_data['crop_names']])
  return user_df
```

```python
user_input = {
  'crop_year': 2003,
  'temperature': 100.5,
  'wind_speed': 70,
  'pressure': 167,
  'humidity': 67,
  'N': 10,
  'P': 9,
  'K': 10,
  'crop_names': 'Jowar' 
}

user_df = prepare_user_input(user_input)
prediction = regr.predict(user_df)[0]  # Assuming model predicts a single value

print("Predicted yield:", prediction)
```
